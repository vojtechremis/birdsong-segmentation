```python
import librosa

# Načtěte MP3 soubor
audio_path = '/Users/vojtechremis/Downloads/XC754598 - Tapajos Fire-eye - Pyriglena similis.mp3'
y, sr = librosa.load(audio_path, sr=None)  # sr=None zachová původní frekvenci vzorkování

print(f"Audio shape: {y.shape}, Sample rate: {sr}")
```

```python
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Vytvoření Mel-spektrogramu
S = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)

# Konverze do decibelové stupnice
S_dB = librosa.power_to_db(S, ref=np.max)

# Vizualizace Mel-spektrogramu
plt.figure(figsize=(10, 4))
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-spectrogram')
plt.tight_layout()
plt.show()
```

```python
# Extrakce MFCC
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

# Vizualizace MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, sr=sr, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```
