```python
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display

# Načtení MP3 souboru
y, sr = librosa.load(file, sr=None)

# Výpočet STFT (Short-Time Fourier Transform)
D_stft = librosa.stft(y)
S_db_stft = librosa.amplitude_to_db(np.abs(D_stft), ref=np.max)

# Výpočet MEL spektrogramu
S_mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, fmax=sr // 2)
S_db_mel = librosa.power_to_db(S_mel, ref=np.max)

# Vizualizace STFT spektrogramu
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
librosa.display.specshow(S_db_stft, sr=sr, x_axis='time', y_axis='log', cmap='magma')
plt.colorbar(format='%+2.0f dB')
plt.title("STFT Spectrogram")

# Vizualizace MEL spektrogramu
plt.subplot(2, 1, 2)
librosa.display.specshow(S_db_mel, sr=sr, x_axis='time', y_axis='mel', cmap='magma')
plt.colorbar(format='%+2.0f dB')
plt.title("MEL Spectrogram")

plt.tight_layout()
plt.show()
 
```
